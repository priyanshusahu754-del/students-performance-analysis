```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# =========================================
# CREATE AVERAGE SCORE
# =========================================

df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3

# Modern theme
sns.set_theme(style="whitegrid")

# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# =========================================
# GENDER VS AVERAGE SCORE
# =========================================

gender_avg = df.groupby("gender")["average_score"].mean().reset_index()

sns.barplot(
    data=gender_avg,
    x="gender",
    y="average_score",
    hue="gender",
    palette="viridis",
    legend=False,
    ax=axes[0]
)

axes[0].set_title("Gender vs Average Score", fontsize=13)
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Average Score")

# Add percentage labels
for i, value in enumerate(gender_avg["average_score"]):
    axes[0].text(
        i,
        value + 1,
        f"{value:.1f}%",
        ha='center',
        fontsize=11
    )

# =========================================
# RACE/ETHNICITY VS AVERAGE SCORE
# =========================================

race_avg = df.groupby("race/ethnicity")["average_score"].mean().reset_index()

sns.barplot(
    data=race_avg,
    x="race/ethnicity",
    y="average_score",
    hue="race/ethnicity",
    palette="magma",
    legend=False,
    ax=axes[1]
)

axes[1].set_title("Race/Ethnicity vs Average Score", fontsize=13)
axes[1].set_xlabel("Group")
axes[1].set_ylabel("Average Score")

# Add percentage labels
for i, value in enumerate(race_avg["average_score"]):
    axes[1].text(
        i,
        value + 1,
        f"{value:.1f}%",
        ha='center',
        fontsize=10
    )

# Adjust spacing
plt.tight_layout()

# Show plots
plt.show()
```

- as we can see that female have better avg score then males
- in different groups we can see that groupA has the lowest scrore where groupE has the highest avg score

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# =========================================
# CREATE AVERAGE SCORE
# =========================================

df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3

# =========================================
# CALCULATE AVG SCORE BY PARENTAL EDUCATION
# =========================================

parent_avg = (
    df.groupby("parental level of education")["average_score"]
    .mean()
    .reset_index()
)

# Sort values
parent_avg = parent_avg.sort_values(
    by="average_score",
    ascending=False
)

# =========================================
# MODERN STYLE
# =========================================

sns.set_theme(style="whitegrid")

# Attractive custom colors
colors = [
    "#6C5CE7",  # Purple
    "#00B894",  # Green
    "#0984E3",  # Blue
    "#E17055",  # Orange
    "#D63031",  # Red
    "#FDCB6E"   # Yellow
]

# =========================================
# GRAPH
# =========================================

plt.figure(figsize=(11,5))

ax = sns.barplot(
    data=parent_avg,
    x="parental level of education",
    y="average_score",
    palette=colors
)

# =========================================
# ADD LABELS
# =========================================

for i, value in enumerate(parent_avg["average_score"]):

    ax.text(
        i,
        value + 1,
        f"{value:.1f}%",
        ha="center",
        fontsize=10,
        fontweight="bold"
    )

# =========================================
# TITLES
# =========================================

plt.title(
    "Parental Education vs Average Score",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel(
    "Parental Level of Education",
    fontsize=12
)

plt.ylabel(
    "Average Score",
    fontsize=12
)

plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

- this plot showing students avg score according to their parents education level 
- in this we can clearly see that parents education is one of the most imp performance deciding factors of student
- Students whose parents have a higher level of education tend to perform better academically.
- Parental education level plays a very important role here because the quality of guidance, learning environment, and academic support at home is usually better when parents are more educated

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# =========================================
# CREATE AVERAGE SCORE
# =========================================

df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3

# =========================================
# CREATE PERFORMANCE CATEGORY
# =========================================

def performance_category(avg):

    if avg >= 80:
        return "Excellent"

    elif avg >= 60:
        return "Good"

    elif avg >= 40:
        return "Average"

    else:
        return "Poor"

df["performance_category"] = df["average_score"].apply(
    performance_category
)

# =========================================
# CONVERT DATA INTO LONG FORMAT
# =========================================

score_data = df.melt(
    id_vars=["gender", "performance_category"],
    value_vars=[
        "math score",
        "reading score",
        "writing score"
    ],
    var_name="Subject",
    value_name="Score"
)

# =========================================
# MODERN STYLE
# =========================================

sns.set_theme(style="whitegrid")

# =========================================
# GRAPH
# =========================================

g = sns.catplot(
    data=score_data,
    x="Subject",
    y="Score",
    hue="gender",
    col="performance_category",
    kind="bar",
    palette=["#6C5CE7", "#00B894"],
    height=4,
    aspect=1
)

# =========================================
# TITLES
# =========================================

g.fig.suptitle(
    "Gender + Subject Scores + Performance Category",
    fontsize=16,
    fontweight="bold"
)

g.set_axis_labels("Subjects", "Average Score")

# =========================================
# ADD PERCENTAGE LABELS
# =========================================

for ax in g.axes.flat:

    for container in ax.containers:

        ax.bar_label(
            container,
            fmt="%.1f%%",
            padding=2,
            fontsize=8
        )

# Adjust spacing
plt.tight_layout()

# Show graph
plt.show()
```

in all category we can clearly see that males are dominationg in maths score whereas in other two subject females are dominating 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# =========================================
# CREATE AVERAGE SCORE
# =========================================

df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3

# =========================================
# SELECT NUMERICAL COLUMNS
# =========================================

numerical_df = df[[
    "math score",
    "reading score",
    "writing score",
    "average_score"
]]

# =========================================
# CORRELATION MATRIX
# =========================================

corr_matrix = numerical_df.corr()

# =========================================
# MODERN STYLE
# =========================================

sns.set_theme(style="white")

# =========================================
# HEATMAP
# =========================================

plt.figure(figsize=(8,5))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=1,
    linecolor="white"
)

# =========================================
# TITLES
# =========================================

plt.title(
    "Correlation Heatmap of Student Scores",
    fontsize=15,
    fontweight="bold"
)

plt.tight_layout()
plt.show()
```

- Reading score aur writing score ke beech bahut strong positive correlation hota hai.
- Jin students ki reading skills achhi hoti hain unki writing skills bhi usually strong hoti hain.
